### Scriviamo media e varianza 

***Estraggo tutte i dati relativi ad epidemie andate  buon fine***

```python
##########################  TRIESTE  #############################################################

S_TS1 = np.array((int(stat[1]*1000),ndays))
E_TS1  = np.array((int(stat[1]*1000),ndays))
Ir_TS1  = np.array((int(stat[1]*1000),ndays))
Iu_TS1  = np.array((int(stat[1]*1000),ndays))
R_TS1  = np.array((int(stat[1]*1000),ndays))



for i in range(0,nrip):
    if S_TS[i,ndays-1] < (2./3.) * S_TS[i,0]:
        S_TS1  = S_TS[i,:]
        E_TS1  = E_TS[i,:]
        Ir_TS1  = Ir_TS[i,:]
        Iu_TS1  = Iu_TS[i,:]
        R_TS1  = R_TS[i,:]
        p = i
        break

for i in range(p + 1, nrip):
    if S_TS[i,ndays-1] < (2./3.) * S_TS[i,0]:
        
        S_TS1  = np.vstack((S_TS1, S_TS[i,:]))
        E_TS1  = np.vstack((E_TS1, E_TS[i,:]))
        Ir_TS1 = np.vstack((Ir_TS1,Ir_TS[i,:]))
        Iu_TS1 = np.vstack((Iu_TS1,Iu_TS[i,:]))
        R_TS1  = np.vstack((R_TS1, R_TS[i,:]))

##########################  UDINE #############################################################

S_UD1 = np.array((int(stat[1]*1000),ndays))
E_UD1  = np.array((int(stat[1]*1000),ndays))
Ir_UD1  = np.array((int(stat[1]*1000),ndays))
Iu_UD1  = np.array((int(stat[1]*1000),ndays))
R_UD1  = np.array((int(stat[1]*1000),ndays))



for i in range(0,nrip):
    if S_UD[i,ndays-1] < (2./3.) * S_UD[i,0]:
        S_UD1   = S_UD[i,:]
        E_UD1   = E_UD[i,:]
        Ir_UD1  = Ir_UD[i,:]
        Iu_UD1  = Iu_UD[i,:]
        R_UD1   = R_UD[i,:]
        p = i
        break

for i in range(p + 1, nrip):
    if S_UD[i,ndays-1] < (2./3.) * S_UD[i,0]:
        
        S_UD1  = np.vstack((S_UD1, S_UD[i,:]))
        E_UD1  = np.vstack((E_UD1, E_UD[i,:]))
        Ir_UD1 = np.vstack((Ir_UD1,Ir_UD[i,:]))
        Iu_UD1 = np.vstack((Iu_UD1,Iu_UD[i,:]))
        R_UD1  = np.vstack((R_UD1, R_UD[i,:]))

########################## GORIZIA #############################################################

S_GO1 = np.array((int(stat[1]*1000),ndays))
E_GO1  = np.array((int(stat[1]*1000),ndays))
Ir_GO1  = np.array((int(stat[1]*1000),ndays))
Iu_GO1  = np.array((int(stat[1]*1000),ndays))
R_GO1  = np.array((int(stat[1]*1000),ndays))



for i in range(0,nrip):
    if S_GO[i,ndays-1] < (2./3.) * S_GO[i,0]:
        S_GO1   = S_GO[i,:]
        E_GO1   = E_GO[i,:]
        Ir_GO1  = Ir_GO[i,:]
        Iu_GO1  = Iu_GO[i,:]
        R_GO1   = R_GO[i,:]
        p = i
        break

for i in range(p + 1, nrip):
    if S_GO[i,ndays-1] < (2./3.) * S_GO[i,0]:
        
        S_GO1  = np.vstack((S_GO1, S_GO[i,:]))
        E_GO1  = np.vstack((E_GO1, E_GO[i,:]))
        Ir_GO1 = np.vstack((Ir_GO1,Ir_GO[i,:]))
        Iu_GO1 = np.vstack((Iu_GO1,Iu_GO[i,:]))
        R_GO1  = np.vstack((R_GO1, R_GO[i,:]))

        
########################## PORDENONE #############################################################

S_PN1 = np.array((int(stat[1]*1000),ndays))
E_PN1  = np.array((int(stat[1]*1000),ndays))
Ir_PN1  = np.array((int(stat[1]*1000),ndays))
Iu_PN1  = np.array((int(stat[1]*1000),ndays))
R_PN1  = np.array((int(stat[1]*1000),ndays))



for i in range(0,nrip):
    if S_PN[i,ndays-1] < (2./3.) * S_PN[i,0]:
        S_PN1   = S_PN[i,:]
        E_PN1   = E_PN[i,:]
        Ir_PN1  = Ir_PN[i,:]
        Iu_PN1  = Iu_PN[i,:]
        R_PN1   = R_PN[i,:]
        p = i
        break

for i in range(p + 1, nrip):
    if S_PN[i,ndays-1] < (2./3.) * S_PN[i,0]:
        
        S_PN1  = np.vstack((S_PN1, S_PN[i,:]))
        E_PN1  = np.vstack((E_PN1, E_PN[i,:]))
        Ir_PN1 = np.vstack((Ir_PN1,Ir_PN[i,:]))
        Iu_PN1 = np.vstack((Iu_PN1,Iu_PN[i,:]))
        R_PN1  = np.vstack((R_PN1, R_PN[i,:]))

        
```

### Calcoliamo Media e Deviazione Standard

```python
########################### TRIESTE  ########################################################à

Sm_TS = np.zeros(len(t))
Em_TS = np.zeros(len(t))
Irm_TS = np.zeros(len(t))
Ium_TS = np.zeros(len(t))
Rm_TS = np.zeros(len(t))
sigmaS_TS  = np.zeros(len(t))
sigmaE_TS  = np.zeros(len(t))
sigmaIr_TS = np.zeros(len(t))
sigmaIu_TS = np.zeros(len(t))
sigmaR_TS  = np.zeros(len(t))


for i in range(0,len(t)):
    Sm_TS[i]  = sum(S_TS1[:,i])/len(S_TS1[:,1])
    Em_TS[i]  = sum(E_TS1[:,i]) / len(S_TS1[:,1])
    Irm_TS[i] = sum(Ir_TS1[:,i])/len(S_TS1[:,1])
    Ium_TS[i] = sum(Iu_TS1[:,i])/len(S_TS1[:,1])
    Rm_TS[i]  = sum(R_TS1[:,i])/len(S_TS1[:,1])
 

for i in range(0,len(t)):
    s_S_TS  = 0.
    s_E_TS  = 0
    s_Ir_TS = 0
    s_Iu_TS  = 0
    s_R_TS  = 0
    
    for j in range(0,len(Sm_TS)):
        s_S_TS  = s_S_TS  + np.power((Sm_TS[i]  - S_TS1[j,i]),  2) 
        s_E_TS  = s_E_TS  + np.power((Em_TS[i]  - E_TS1[j,i]),  2) 
        s_Ir_TS = s_Ir_TS + np.power((Irm_TS[i] - Ir_TS1[j,i]), 2) 
        s_Iu_TS = s_Iu_TS + np.power((Ium_TS[i] - Iu_TS1[j,i]), 2) 
        s_R_TS  = s_R_TS  + np.power((Rm_TS[i]  - R_TS1[j,i]),  2)  
        
    sigmaS_TS[i]  = np.sqrt(s_S_TS/len(S_TS1[:,1]))
    sigmaE_TS[i]  = np.sqrt( s_E_TS /len(S_TS1[:,1]))
    sigmaIr_TS[i] = np.sqrt(s_Ir_TS/len(S_TS1[:,1]))
    sigmaIu_TS[i] = np.sqrt(s_Iu_TS/len(S_TS1[:,1]))
    sigmaR_TS[i]  = np.sqrt(s_R_TS/len(S_TS1[:,1]))
    
    ########################### UDINE  ########################################################à

Sm_UD = np.zeros(len(t))
Em_UD = np.zeros(len(t))
Irm_UD = np.zeros(len(t))
Ium_UD = np.zeros(len(t))
Rm_UD = np.zeros(len(t))
sigmaS_UD  = np.zeros(len(t))
sigmaE_UD = np.zeros(len(t))
sigmaIr_UD = np.zeros(len(t))
sigmaIu_UD = np.zeros(len(t))
sigmaR_UD  = np.zeros(len(t))

for i in range(0,len(t)):
    Sm_UD[i]  = sum(S_UD1[:,i])/len(S_UD1[:,1])
    Em_UD[i]  = sum(E_UD1[:,i])/len(S_UD1[:,1])
    Irm_UD[i] = sum(Ir_UD1[:,i])/len(S_UD1[:,1])
    Ium_UD[i] = sum(Iu_UD1[:,i])/len(S_UD1[:,1])
    Rm_UD[i]  = sum(R_UD1[:,i])/len(S_UD1[:,1])
 

for i in range(0,len(t)):
    s_S_UD  = 0.
    s_E_UD  = 0
    s_Ir_UD = 0
    s_Iu_UD  = 0
    s_R_UD  = 0
    
    for j in range(0,len(Sm_UD)):
        s_S_UD  = s_S_UD  + np.power((Sm_UD[i]  - S_UD1[j,i]),  2) 
        s_E_UD  = s_E_UD  + np.power((Em_UD[i]  - E_UD1[j,i]),  2) 
        s_Ir_UD = s_Ir_UD + np.power((Irm_UD[i] - Ir_UD1[j,i]), 2) 
        s_Iu_UD = s_Iu_UD + np.power((Ium_UD[i] - Iu_UD1[j,i]), 2) 
        s_R_UD  = s_R_UD  + np.power((Rm_UD[i]  - R_UD1[j,i]),  2)  
        
    sigmaS_UD[i]  = np.sqrt(s_S_UD/len(S_UD1[:,1]))
    sigmaE_UD[i]  = np.sqrt(s_E_UD/len(S_UD1[:,1]))
    sigmaIr_UD[i] = np.sqrt(s_Ir_UD/len(S_UD1[:,1]))
    sigmaIu_UD[i] = np.sqrt(s_Iu_UD/len(S_UD1[:,1]))
    sigmaR_UD[i]  = np.sqrt(s_R_UD/len(S_UD1[:,1]))
    
    ########################### GORIZIA ########################################################à

Sm_GO = np.zeros(len(t))
Em_GO = np.zeros(len(t))
Irm_GO = np.zeros(len(t))
Ium_GO = np.zeros(len(t))
Rm_GO = np.zeros(len(t))
sigmaS_GO = np.zeros(len(t))
sigmaE_GO  = np.zeros(len(t))
sigmaIr_GO = np.zeros(len(t))
sigmaIu_GO = np.zeros(len(t))
sigmaR_GO  = np.zeros(len(t))

for i in range(0,len(t)):
    Sm_GO[i]  = sum(S_GO1[:,i])/len(S_GO1[:,1])
    Em_GO[i]  = sum(E_GO1[:,i])/len(S_GO1[:,1])
    Irm_GO[i] = sum(Ir_GO1[:,i])/len(S_GO1[:,1])
    Ium_GO[i] = sum(Iu_GO1[:,i])/len(S_GO1[:,1])
    Rm_GO[i]  = sum(R_GO1[:,i])/len(S_GO1[:,1])
 

for i in range(0,len(t)):
    s_S_GO  = 0.
    s_E_GO  = 0
    s_Ir_GO = 0
    s_Iu_GO = 0
    s_R_GO  = 0
    
    for j in range(0,len(Sm_GO)):
        s_S_GO  = s_S_GO  + np.power((Sm_GO[i]  - S_GO1[j,i]),  2) 
        s_E_GO  = s_E_GO  + np.power((Em_GO[i]  - E_GO1[j,i]),  2) 
        s_Ir_GO = s_Ir_GO + np.power((Irm_GO[i] - Ir_GO1[j,i]), 2) 
        s_Iu_GO = s_Iu_GO + np.power((Ium_GO[i] - Iu_GO1[j,i]), 2) 
        s_R_GO  = s_R_GO  + np.power((Rm_GO[i]  - R_GO1[j,i]),  2)  
        
    sigmaS_GO[i]  = np.sqrt(s_S_GO/len(S_GO1[:,1]))
    sigmaE_GO[i]  = np.sqrt(s_E_GO/len(S_GO1[:,1]))
    sigmaIr_GO[i] = np.sqrt(s_Ir_GO/len(S_GO1[:,1]))
    sigmaIu_GO[i] = np.sqrt(s_Iu_GO/len(S_GO1[:,1]))
    sigmaR_GO[i]  = np.sqrt(s_R_GO/len(S_GO1[:,1]))
    
    ########################### PORDENONE  ########################################################à

Sm_PN = np.zeros(len(t))
Em_PN = np.zeros(len(t))
Irm_PN = np.zeros(len(t))
Ium_PN = np.zeros(len(t))
Rm_PN = np.zeros(len(t))
sigmaS_PN  = np.zeros(len(t))
sigmaE_PN  = np.zeros(len(t))
sigmaIr_PN = np.zeros(len(t))
sigmaIu_PN = np.zeros(len(t))
sigmaR_PN  = np.zeros(len(t))

for i in range(0,len(t)):
    Sm_PN[i]  = sum(S_PN1[:,i])/len(S_PN1[:,1])
    Em_PN[i]  = sum(E_PN1[:,i])/len(S_PN1[:,1])
    Irm_PN[i] = sum(Ir_PN1[:,i])/len(S_PN1[:,1])
    Ium_PN[i] = sum(Iu_PN1[:,i])/len(S_PN1[:,1])
    Rm_PN[i]  = sum(R_PN1[:,i])/len(S_PN1[:,1])
 

for i in range(0,len(t)):
    s_S_PN  = 0.
    s_E_PN  = 0
    s_Ir_PN = 0
    s_Iu_PN  = 0
    s_R_PN  = 0
    
    for j in range(0,len(Sm_PN)):
        s_S_PN  = s_S_PN  + np.power((Sm_PN[i]  - S_PN1[j,i]),  2) 
        s_E_PN  = s_E_PN  + np.power((Em_PN[i]  - E_PN1[j,i]),  2) 
        s_Ir_PN = s_Ir_PN + np.power((Irm_PN[i] - Ir_PN1[j,i]), 2) 
        s_Iu_PN = s_Iu_PN + np.power((Ium_PN[i] - Iu_PN1[j,i]), 2) 
        s_R_PN  = s_R_PN  + np.power((Rm_PN[i]  - R_PN1[j,i]),  2)  
        
    sigmaS_PN[i]  = np.sqrt(s_S_PN/len(S_PN1[:,1]))
    sigmaE_PN[i]  = np.sqrt(s_E_PN/len(S_PN1[:,1]))
    sigmaIr_PN[i] = np.sqrt(s_Ir_PN/len(S_PN1[:,1]))
    sigmaIu_PN[i] = np.sqrt(s_Iu_PN/len(S_PN1[:,1]))
    sigmaR_PN[i]  = np.sqrt(s_R_PN/len(S_PN1[:,1]))
    
```
