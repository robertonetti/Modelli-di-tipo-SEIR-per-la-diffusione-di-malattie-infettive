# Grafici blocco mobilità

```python
import numpy as np                                  # libreria matematica
import scipy as sp
import matplotlib.pyplot as plt                     # librerie grafiche
import matplotlib.colors as mcolors
import pandas as pd
plt.rcParams.update({'font.size': 22})
```

### Estrazione statistiche

```python


s_10  = pd.read_csv('stat10', names=['stat'], delim_whitespace=True).to_numpy()
s_20  = pd.read_csv('stat20', names=['stat'], delim_whitespace=True).to_numpy()
s_30  = pd.read_csv('stat30', names=['stat'], delim_whitespace=True).to_numpy()
s_40  = pd.read_csv('stat40', names=['stat'], delim_whitespace=True).to_numpy()
s_50  = pd.read_csv('stat50', names=['stat'], delim_whitespace=True).to_numpy()

print(s_10[3][0])
```

```python

UD = s_10[1][0]
GO = s_10[2][0]
PN = s_10[3][0]

UD = np.append(UD,[s_20[1][0],s_30[1][0],s_40[1][0],s_50[1][0]])
GO = np.append(GO,[s_20[2][0],s_30[2][0],s_40[2][0],s_50[2][0]])
PN = np.append(PN,[s_20[3][0],s_30[3][0],s_40[3][0],s_50[3][0]])

UD = UD*100
GO = GO*100
PN = PN*100

```

### Grafico probabilità estinzione dovuta al blocco della mobilità

```python
x = np.array([10,20,30,40,50])

plt.figure(figsize=(20,20))

plt.plot(x,UD,'b',linewidth=3.,marker="o",markersize= 10.,label='Udine')
plt.plot(x,GO,'g',linewidth=3.,marker="o",markersize= 10.,label='Gorizia')
plt.plot(x,PN,'k',linewidth=3.,marker="o",markersize= 10.,label='Pordenone')
plt.xlabel('NUMERO INFETTI REPORTED DI SOGLIA')
plt.ylabel('% EPIDEMIE SCAMPATE')
plt.title('STATISTICHE EPIDEMIE')
plt.legend(title="Diffusione del virus in:")
plt.grid(True)
plt.savefig('statistiche_intervention.png')
```
