# ESTRAZIONE DATI MATRICE PENDOLARISMO

 Importo delle librerie utili

```python
import numpy as np                  #libreria numerica
import pandas as pd                 #libreria manipolazione dati
```

Carico la matrice del pendolarismo italiana

```python
ITA=pd.read_csv('matrix_pendo2011_10112014.txt',delim_whitespace=True, 
               names=['L/S','tipo_residenza','provincia_residenza','comune_residenza','sesso',
                      'motivo','luogo','provincia_spostamento','comune_spostamento','stato estero',
                      'mezzo','orario','tempo','stima','numero'])
```

Estraggo la matrice del pendolarismo Fvg

```python
#filtro per i viaggi tra qualsiasi regione e il Friuli

filter1 = [30,31,32,93] #udine, gorizia, trieste, pordenone

FVG_andata = ITA.query('provincia_residenza == @filter1 ')   #dal friuli a qualsiasi regione
FVG_arrivo = ITA.query('provincia_spostamento == @filter1 ') #da qualsiasi regione al friuli

# filtro da Fvg ad Fvg esclusivamente

FVG = ITA.query('provincia_residenza == @filter1 and provincia_spostamento == @filter1');
FVG
```

## Definisco la matrice Mij a partire da FVG

### Legenda delle cittÃ 

**Trieste**   = 1,                 provincia 32, cittÃ  6


**Udine**     = 2,     provincia 30, cittÃ  129


**Gorizia**   = 3,     provincia 31, cittÃ  7


**Pordenone** = 4,     provincia 93, cittÃ  33

### Costruisco matrici CittÃ 1-CittÃ 2

```python
A1_2 = FVG.query('provincia_residenza == 32 and comune_residenza == 6 and provincia_spostamento == 30 and comune_spostamento == 129');
A1_3 = FVG.query('provincia_residenza == 32 and comune_residenza == 6 and provincia_spostamento == 31 and comune_spostamento == 7')
A1_4 = FVG.query('provincia_residenza == 32 and comune_residenza == 6 and provincia_spostamento == 93 and comune_spostamento == 33')
A2_1 = FVG.query('provincia_residenza == 30 and comune_residenza == 129 and provincia_spostamento == 32 and comune_spostamento == 6')
A2_3 = FVG.query('provincia_residenza == 30 and comune_residenza == 129 and provincia_spostamento == 31 and comune_spostamento == 7')
A2_4 = FVG.query('provincia_residenza == 30 and comune_residenza == 129 and provincia_spostamento == 93 and comune_spostamento == 33')
A3_2 = FVG.query('provincia_residenza == 31 and comune_residenza == 7 and provincia_spostamento == 30 and comune_spostamento == 129')
A3_1 = FVG.query('provincia_residenza == 31 and comune_residenza == 7 and provincia_spostamento == 32 and comune_spostamento == 6')
A3_4 = FVG.query('provincia_residenza == 31 and comune_residenza == 7 and provincia_spostamento == 93 and comune_spostamento == 33')
A4_2 = FVG.query('provincia_residenza == 93 and comune_residenza == 33 and provincia_spostamento == 30 and comune_spostamento == 129')
A4_3 = FVG.query('provincia_residenza == 93 and comune_residenza == 33 and provincia_spostamento == 31 and comune_spostamento == 7')
A4_1 = FVG.query('provincia_residenza == 93 and comune_residenza == 33 and provincia_spostamento == 32 and comune_spostamento == 6')
A1_2
```

### Costruisco Mij

```python

M = np.zeros((5,5));
M[1][2]= float(sum(A1_2.agg({'stima':'sum'})))
M[1][3]= float(sum(A1_3.agg({'stima':'sum'})))
M[1][4]= float(sum(A1_4.agg({'stima':'sum'})))
M[2][1]= float(sum(A2_1.agg({'stima':'sum'})))
M[2][3]= float(sum(A2_3.agg({'stima':'sum'})))
M[2][4]= float(sum(A2_4.agg({'stima':'sum'})))
M[3][1]= float(sum(A3_1.agg({'stima':'sum'})))
M[3][2]= float(sum(A3_2.agg({'stima':'sum'})))
M[3][4]= float(sum(A3_4.agg({'stima':'sum'})))
M[4][1]= float(sum(A4_1.agg({'stima':'sum'})))
M[4][2]= float(sum(A4_2.agg({'stima':'sum'})))
M[4][3]= float(sum(A4_3.agg({'stima':'sum'})))

# salvo un file Mfvg
np.savetxt("Mfvg.txt", M, delimiter=",")
 
print(M)
```
