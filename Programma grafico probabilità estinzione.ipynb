```python
TS = TS_1[1][0]
UD = UD_1[1][0]
GO = GO_1[1][0]
PN = PN_1[1][0]

TS = np.append(TS,[TS_2[1][0],TS_3[1][0],TS_4[1][0],TS_5[1][0],TS_6[1][0],TS_7[1][0],TS_8[1][0]])
UD = np.append(UD,[UD_2[1][0],UD_3[1][0],UD_4[1][0],UD_5[1][0],UD_6[1][0],UD_7[1][0],UD_8[1][0]])
GO = np.append(GO,[GO_2[1][0],GO_3[1][0],GO_4[1][0],GO_5[1][0],GO_6[1][0],GO_7[1][0],GO_8[1][0]])
PN = np.append(PN,[PN_2[1][0],PN_3[1][0],PN_4[1][0],PN_5[1][0],PN_6[1][0],PN_7[1][0],PN_8[1][0]])
TS = TS*100
UD = UD*100
GO = GO*100
PN = PN*100
```

## Grafico percentuale di epidemie estinte

```python
x = np.array([1,2,3,4,5,6,7,8])

plt.figure(figsize=(20,20))
plt.plot(x,TS,'r',linewidth=3.,marker="o",markersize= 10.,label='Trieste')
plt.plot(x,UD,'b',linewidth=3.,marker="o",markersize= 10.,label='Udine')
plt.plot(x,GO,'g',linewidth=3.,marker="o",markersize= 10.,label='Gorizia')
plt.plot(x,PN,'k',linewidth=3.,marker="o",markersize= 10.,label='Pordenone')
plt.xlabel('NUMERO INFETTI UNDOCUMENTED INIZIALI')
plt.ylabel('% EPIDEMIE SCAMPATE')
plt.title('STATISTICHE EPIDEMIE')
plt.legend(title="Epidemia parte da:")
plt.grid(True)
plt.savefig('andamenti statistici.png')
```
